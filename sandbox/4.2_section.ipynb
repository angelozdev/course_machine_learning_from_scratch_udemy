```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import math
```

```python
X_1 = math.pi * np.random.rand(100, 1)
X_2 = math.e * np.random.rand(100, 1)

y = X_1 * 3 + X_2 * 2 + np.random.randn(100, 1)

scaler = StandardScaler()

X_1_scaled = scaler.fit_transform(X_1)
X_2_scaled = scaler.fit_transform(X_2)

X_scaled = np.hstack((X_1_scaled, X_2_scaled))

plt.scatter(X_1_scaled, y, label="X_1")
plt.scatter(X_2_scaled, y, label="X_2")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Datos Sintéticos para Regresión")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=1997)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_pred = lin_reg.predict(X_test)

residuals = y_test - y_pred

print("Mean error: ", np.mean(residuals))
print("Mean Absolute Error (MAE)", np.mean(np.abs(residuals)))
print("Mean Squared Error (MSE)", np.mean(residuals**2))
print("Root Mean Squared Error (RMSE)", np.sqrt(np.mean(residuals**2)))
print("Std error: ", np.std(residuals))
print("Mean error / Std error: ", np.mean(residuals) / np.std(residuals))
print("R^2: ", lin_reg.score(X_test, y_test))
print("X_train Max: ", np.max(X_train))
print("X_train Min: ", np.min(X_train))
print("X_test Max: ", np.max(X_test))
print("X_test Min: ", np.min(X_test))
print("y_train Max: ", np.max(y_train))
print("y_train Min: ", np.min(y_train))
print("y_test Max: ", np.max(y_test))
print("y_test Min: ", np.min(y_test))
print("y_pred Max: ", np.max(y_pred))
print("y_pred Min: ", np.min(y_pred))
print("residuals Max: ", np.max(residuals))
print("residuals Min: ", np.min(residuals))
print("X_1 Max: ", np.max(X_1))
print("X_1 Min: ", np.min(X_1))
print("X_2 Max: ", np.max(X_2))
print("X_2 Min: ", np.min(X_2))
print("y Max: ", np.max(y))
print("y Min: ", np.min(y))

plt.scatter(y_pred, residuals, alpha=0.7)
plt.xlabel("Predicción")
plt.ylabel("Residuos")
plt.title("Residuos vs. Predicción (Test)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.axhline(y=0, color="red", linestyle="--")
plt.axhline(y=min(residuals), color="blue", linestyle="-", label="Min residual")
plt.axhline(y=max(residuals), color="blue", linestyle="-", label="Max residual")
plt.legend()
plt.show()
```
